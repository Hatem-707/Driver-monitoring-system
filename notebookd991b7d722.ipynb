```python
import os
if not os.path.isdir("/kaggle/working/dataset_formated"):
    os.mkdir("/kaggle/working/dataset_formated")

if (not (os.path.isdir("/kaggle/working/dataset_formated/train"))):
    os.mkdir("/kaggle/working/dataset_formated/train")
    os.mkdir("/kaggle/working/dataset_formated/train/Drowsy")
    os.mkdir("/kaggle/working/dataset_formated/train/Non Drowsy")
if (not os.path.isdir("/kaggle/working/dataset_formated/test")):
    os.mkdir("/kaggle/working/dataset_formated/test")
    os.mkdir("/kaggle/working/dataset_formated/test/Drowsy")
    os.mkdir("/kaggle/working/dataset_formated/test/Non Drowsy")

for dirpath, _, filenames in os.walk("/kaggle/working/dataset"):
    length = len(filenames)
    for i, filename in enumerate(filenames):
        full_path = os.path.join(dirpath, filename)
        new_path = ""
        if i<0.8*length:
            new_path = full_path.replace("dataset", "dataset_formated/train")
        else:
            new_path = full_path.replace("dataset", "dataset_formated/test")
        os.replace(full_path, new_path)
        
```

```python
if (not os.path.isdir("/kaggle/working/dataset_formated/val")):
    os.mkdir("/kaggle/working/dataset_formated/val")
    os.mkdir("/kaggle/working/dataset_formated/val/Drowsy")
    os.mkdir("/kaggle/working/dataset_formated/val/Non Drowsy")
for dirpath, _, filenames in os.walk("/kaggle/working/dataset_formated/test"):
    length = len(filenames)
    for i, filename in enumerate(filenames):
        full_path = os.path.join(dirpath, filename)
        new_path = ""
        if i<0.5*length:
            new_path = full_path.replace("test", "val")
            os.replace(full_path, new_path)
        else:
            continue
```
